```python
from natal import Data, Stats, Chart, Config
from datetime import datetime
from IPython.display import HTML
```

```python
# Birth Chart

mimi = Data(
    name = "MiMi",
    city = "Taipei",
    dt = "1980-04-20 14:30",
    config = Config(theme_type="light")
)

chart = Chart(data1=mimi, width=600)
HTML(chart.svg)
```

```python
# transit chart
from natal.config import Orb

mimi.config.theme_type="dark"
mimi.config.orb = Orb(conjunction=2, opposition=2, square=2, trine=2, sextile=1)

current = Data(name="Current", city="Taipei", dt=datetime.now())

chart = Chart(data1=mimi, data2=current, width=600)
HTML(chart.svg)
```

```python
# Statistics of Celestial Bodies

stats = Stats(data1=mimi, data2=current)
HTML(stats.full_report("html"))
```

```python
# Monochrome Chart
mimi.config.theme_type = "mono"
chart = Chart(data1=mimi, width=600)
HTML(chart.svg)
```
