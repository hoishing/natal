```python
from IPython.display import HTML
from natal import Chart, Config, Data
```

```python
# Birth Chart

mimi = Data(
    name="MiMi",
    utc_dt="1980-04-20 06:30:00",
    lat=25.0531,
    lon=121.526,
)

chart = Chart(data1=mimi, width=600)
HTML(chart.svg)
```

```python
# transit chart

from natal.config import Orb

mimi.config = Config(
    orb=Orb(conjunction=2, opposition=2, trine=2, square=2, sextile=1),
)

transit = Data(
    name="Transit",
    utc_dt="2024-01-01 05:30:00",
    lat=25.0531,
    lon=121.526,
)

chart = Chart(data1=mimi, data2=transit, width=600)
HTML(chart.svg)
```

```python
# Light theme with asteroids

from natal.config import Display

mimi.config = Config(
    theme_type="light",
    display=Display(
        south_node=True,
        chiron=True,
        ceres=True,
        pallas=True,
        juno=True,
        vesta=True,
    ),
)
chart = Chart(data1=mimi, width=600)
HTML(chart.svg)
```

```python
# Monochrome Chart

mimi.config.theme_type = "mono"
chart = Chart(data1=mimi, width=600)
HTML(chart.svg)
```
