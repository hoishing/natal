```python
from natal import Data, Stats, Chart, Config
from datetime import datetime
from IPython.display import HTML
```

```python
# Birth Chart

mimi = Data(
    name = "MiMi",
    city = "Taipei",
    dt = "1980-04-20 14:30"
)

config = Config(theme_type="light")

chart = Chart(data1=mimi, width=600, config=config)
HTML(chart.svg)
```

```python
# transit chart

current = Data(name="Current", city="Taipei", dt=datetime.now())

transit_config = Config(
    theme_type="dark",
    orb={"conjunction": 2, "opposition": 2, "square": 2, "trine": 2, "sextile": 1},
)

chart = Chart(data1=mimi, data2=current, width=600, config=transit_config)
HTML(chart.svg)
```

```python
# Statistics of Celestial Bodies

mimi.config = transit_config
stats = Stats(data1=mimi, data2=current)
HTML(stats.full_report("html"))
```

```python
# Monochrome Chart

chart = Chart(data1=mimi, width=600, config=Config(theme_type="mono"))
HTML(chart.svg)
```
